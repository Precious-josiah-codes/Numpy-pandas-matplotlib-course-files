Operations On DataFrames and the Aggregation functions*

sum: Return the sum of the values for the requested axis
min: Return the minimum of the values for the requested axis
max: Return the maximum of the values for the requested axis

syntax:
dataframe.aggregate(func, axis, args, kwargs)

axis	0
1
'index'
'columns'	Optional, Which axis to apply the function to. default 0.

```python
data = {
    'x': [1,1,1],
    'y': [300, 1112, 42]
}

df = pd.DataFrame(data)
print(df)

# sum by rows
x = df.aggregate(['sum'], axis=1)
x

# sum by columns
x = df.aggregate(['sum'], axis=0)
x

# max
x = df.aggregate(['max'], axis=0)
x
```

Boolean DataFrame and Masking
Pandas provides a feature called Boolean Masks that let's you filter DataFrames based on conditions. With this, we can write simple queries to filter our data.

```python
df = pd.DataFrame([
	{
		"person": "James",
		"sales": 1000,
		"lastName": "Taylor",
	},
	{
		"person": "Luna",
		"sales": 2000,
		"lastName": "Mound"
	},
	{
		"person": "Clara",
		"sales": 3000,
		"lastName": "Brown"
	}
])
df


# filtered = df[df['sales'] > 1000]
# filtered

# filter rows with multiple boolean mask
# using | for or in python as well as & for and in python
filtered = df[(df['sales'] <= 1000) | (df['sales'] >= 3000)]
filtered

# using &
filtered = df[(df['sales'] > 1000) & (df['sales'] < 3000)]
filtered
```
