Operations On DataFrames and the Aggregation functions*

sum: Return the sum of the values for the requested axis
min: Return the minimum of the values for the requested axis
max: Return the maximum of the values for the requested axis

syntax:
dataframe.aggregate(func, axis, args, kwargs)

axis	0
1
'index'
'columns'	Optional, Which axis to apply the function to. default 0.

```python
data = {
    'x': [1,1,1],
    'y': [300, 1112, 42]
}

df = pd.DataFrame(data)
print(df)

# sum by rows
x = df.aggregate(['sum'], axis=1)
x

# sum by columns
x = df.aggregate(['sum'], axis=0)
x

# max
x = df.aggregate(['max'], axis=0)
x
```

Boolean DataFrame and Masking
Pandas provides a feature called Boolean Masks that let's you filter DataFrames based on conditions. With this, we can write simple queries to filter our data.

```python
df = pd.DataFrame([
	{
		"person": "James",
		"sales": 1000,
		"lastName": "Taylor",
	},
	{
		"person": "Luna",
		"sales": 2000,
		"lastName": "Mound"
	},
	{
		"person": "Clara",
		"sales": 3000,
		"lastName": "Brown"
	}
])
df


# filtered = df[df['sales'] > 1000]
# filtered

# filter rows with multiple boolean mask
# using | for or in python as well as & for and in python
filtered = df[(df['sales'] <= 1000) | (df['sales'] >= 3000)]
filtered

# using &
filtered = df[(df['sales'] > 1000) & (df['sales'] < 3000)]
filtered
```

What is scaling in pandas?
Scaling is an operation where you change the range (i.e. min-max) of a dataframe. Its useful for setting different parameters in the same scale and is also necessary for many algorithms

There are two most common techniques of how to scale columns of Pandas dataframe â€“ Min-Max Normalization and Standardization.

X_{norm} = (X-X_{min})/(X_{max}-X_{min})

```python
# Min-Max Normalization
# Here, all the values are scaled in between the range of[0, 1] where 0 is the minimum value and 1 is the maximum value. The formula for Min-Max Normalization is

# Method 1: using pandas and numpy
df = pd.read_csv('data.csv')
df
new_df = df['Duration'].values.astype(float)
new_df
df_norm = (new_df-new_df.min()) / (new_df.max()-new_df.min())
df_norm



```

```python
# Method 2: using MinMaxScaler from sklearn
from sklearn.preprocessing import MinMaxScaler
df = pd.read_csv('data.csv')
scaler = MinMaxScaler()

df_scales = scaler.fit_transform(df.to_numpy())
df_scales
df_scaled = pd.DataFrame(df_scales, columns=['Duration', 'Pulse', 'Maxpulse', 'Calories'])

# print('Scaled dataset uisng Minmaxscaler')
df_scaled
```

Standardization:
Standardization doesnâ€™t have any fixed minimum or maximum value. Here, the values of all the columns are scaled in such a way that they all have a mean equal to 0 and standard deviation equal to 1

```python
# Method 2: using Standard scaler from sklearn
from sklearn.preprocessing import StandardScaler
df = pd.read_csv('data.csv')
scaler = StandardScaler()

df_scales = scaler.fit_transform(df.to_numpy())
df_scales
df_scaled = pd.DataFrame(
    df_scales, columns=['Duration', 'Pulse', 'Maxpulse', 'Calories'])

# print('Scaled dataset uisng Minmaxscaler')
df_scaled
```

Appending New Rows to a dataFrame

dataframe.append() function is used to append rows of other dataframe to the end of the given dataframe, returning a new dataframe object. Columns not in the original dataframes are added as new columns and the new cells are populated with NaN value.

```python
df1 = pd.DataFrame({
    'a': [1,2,3,4],
    'b': [5,6,7,8]
})

df2 = pd.DataFrame({
    'a': [10, 10, 11, 12],
    'b': [82, 62, 7, 8]
})

# to append df2 at the end of df1 dataframe
df1.append(df2)
```
