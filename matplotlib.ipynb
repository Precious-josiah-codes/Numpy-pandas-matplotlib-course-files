```python
# PLOT
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,8])
ypoints = np.array([3,10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
# BARCHART
# DRAW 4 BARS
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3,9,1,10])


plt.bar(x,y)
plt.show()
```

```python
# PIE CHART
y = np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

```python
# HISTOGRAM
# AS A GRAPH THAT SHOWS THE NUMBER OF OBSERVATIONS WITHIN EACH GIVEN INTERVAL
# USE TO SHOW FREQUENCY DISTRIBUTION

# normal = 80 -100
# medium = 113 - 129
# high = 130 - 150

blood_sugar = [113,85,90,150,149,88,93,115,80,77,82,129]
plt.hist(blood_sugar, bins=3, rwidth=0.95)
plt.show()
```

```python
# SCATTER PLOT
# the scatter plots one dot for each observation. it needs two arrays of same length, one for the values of the x-axis, and one for the values on the y-axis

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x,y)
plt.show()
```

```python
# STACK PLOT
days = [1,2,3,4,5]

sleep = [6,7,5,8,6]
eat = [2,2,1,2,1]
work = [5,7,10,8,6]
exercise = [3,3,0,1,3]
plt.stackplot(days, sleep, eat, work, exercise, colors=['indigo', 'pink', 'red', 'blue'])
plt.show()
```

Legend, Labels, title, xticks and grid

```python
# STACK BAR CHART

city = ['Delhi', 'Beijeing', 'Washington', 'Tokyo', 'Moscow']
gender = ['male', 'female']
pos = np.arange(len(city))
happiness_index_male = [60, 40, 70, 65, 85]
happiness_index_female = [30, 60, 70, 55, 75]

plt.bar(pos, happiness_index_male, color='blue')
plt.bar(pos, happiness_index_female, color='pink', bottom=happiness_index_male)

# legend
plt.legend(gender)

# labels
plt.xlabel('City')
plt.ylabel('happiness index')

# xticks
plt.xticks(pos, city)

# title
plt.title('Stacked bar chart - Happiness index across cities')
# plt.title('Stacked bar chart - Happiness index across cities', fontsize=16)

plt.show()

```

GRID

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240, 250,260,270,280,290,300,310,320,330])

plt.title('Sports Watch Data')
plt.xlabel('Scores')
plt.ylabel('time')

plt.plot(x, y)
# plt.grid()

# specify the axis you want the grid to show
plt.grid(axis='x')
```

```python
ypoints = np.array([3,8,1,10])
plt.plot(ypoints, marker='D')
plt.show()

# LIST OF MARKERS
# 'o'	Circle
# '*'	Star
# '.'	Point
# ','	Pixel
# 'x'	X
# 'X'	X(filled)
# '+'	Plus
# 'P'	Plus(filled)
# 's'	Square
# 'D'	Diamond
# 'd'	Diamond(thin)
# 'p'	Pentagon
# 'H'	Hexagon
# 'h'	Hexagon
# 'v'	Triangle Down
# '^'	Triangle Up
# '<'	Triangle Left
# '>'	Triangle Right
# '1'	Tri Down
# '2'	Tri Up
# '3'	Tri Left
# '4'	Tri Right
# '|'	Vline
# '_'	Hline
```

Multiple chart on same axis

```python
# STACK BAR CHART



city = ['Delhi', 'Beijeing', 'Washington', 'Tokyo', 'Moscow']
gender = ['male', 'female']
pos = np.arange(len(city))
bar_width = 0.35
happiness_index_male = [60, 40, 70, 65, 85]
happiness_index_female = [30, 60, 70, 55, 75]

plt.bar(pos, happiness_index_male, bar_width, color='blue')
plt.bar(pos + bar_width, happiness_index_female, bar_width, color='pink')



# # legend
plt.legend(gender)

# # labels
# plt.xlabel('City')
plt.ylabel('happiness index')

# # xticks
plt.xticks(pos, city)

# title
plt.title('Stacked bar chart - Happiness index across cities')
# plt.title('Stacked bar chart - Happiness index across cities', fontsize=16)
# plt.grid()
plt.show()
```

Subplots
with subplots you can draw multiple plots in one figure

```python
# plot 1
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.title('SALES')
plt.plot(x,y)
#  1 = row
#  2 = column
#  1 = index

x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.title('INCOME')
plt.suptitle('MY SHOP')
plt.plot(x,y)


plt.show()
```
